This script aims to sample subset of posts and manually annotate them in 3 dimensions--intervention, medical consensus and treatment. -1 means Yes, 0 means Neutral, and 1 means No for each dimension.

As mentioned in *Reddit_scrapy.ipynb*, we need a refinement in posts to filter out those less relevant posts.

```python
import pandas as pd
```

```python
df = pd.read_csv('data/posts_en.csv')
print(f"Number of posts: {len(df)}")
```

Next, we define a keyword list derived from `searching_keywords` and filter those posts without these words

可能出问题了，之前检索的时候没有这么少啊，估计要删掉这部分 

```python
search_keywords = [
    "depression", 
    "depressed", 
    "depressive disorder", 
    "major depressive disorder", 
    "mdd",
    "antidepressants",
    "ssri", 
    "snri", 
    "CBT", 
    "DBT",
    "psychotherapy", 
    "therapy", 
    "medication",
    "treatment"
]

# examine whether title and text contain keyword
def contains_keyword(row):
    text = f"{row['Title']} {row['Sentence_normalized']}".lower()
    return any(kw.lower() in text for kw in search_keywords)

# keep posts containing keywords
filtered_df = df[df.apply(contains_keyword, axis=1)]
filtered_df.to_csv("data/posts_en_filter.csv")

print(f"Number of posts after keyword filtering: {len(filtered_df)}")
```

```python
# randomly sample 500 posts
top_posts = filtered_df.sort_values(by='Total_comments', ascending=False).head(1000)

sample_posts = top_posts.sample(n=500, random_state=42)

# add empty column
sample_posts['Information on Interventions'] = ''
sample_posts['Alignment with Medical Consensus'] = ''
sample_posts['Evidence-based Treatment'] = ''
sample_posts['Note'] = ''

# save to csv
sample_posts.to_csv('data/annotation_sample.csv', encoding='utf-8', index=False)

print("Annotation sample saved as 'annotation_sample.csv'")
```
