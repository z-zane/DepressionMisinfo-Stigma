```python
from LexNorm import Normalizer
import contractions
```

```python
def test_cases():
    normalizer = Normalizer()
    test_texts = [
        "lol",
        "LOL",
        "Lol!",
        "I am lol at this.",
        "iving in the city.",
        "My bff is 4ever.",
        "lmgdao that's so funny",
        "I recieved the letter.",
        "teh quick brown fox.",
        "I wanna have meds",
        "I gotta eat",
        "It is kinda little",
        "I'm gonna go asap.",
        "fr bro I support it",
        "everything's ok",
        "It's been ok",
        "I'm ftm"
    ]
    for text in test_texts:
        # 归一化流程
        expanded_text = contractions.fix(text)
        norm_tokens = normalizer.normalize([expanded_text])
        n_text, total_cnt, replaced, replaced_with, spelling_corrections = normalizer.correct_spelling_mistakes(norm_tokens)
        norm_sentence = " ".join(n_text[0])
        print(f"Original: {text}\nNormalized: {norm_sentence}\n")
```

```python
if __name__ == "__main__":
    test_cases()
```
