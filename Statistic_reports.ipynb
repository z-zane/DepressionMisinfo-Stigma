```python
# DPMisinfo Dataset
import pandas as pd

df = pd.read_csv("data/comments_en.csv")

min_time = df["Timestamp"].min()
max_time = df["Timestamp"].max()

print(f"earliest time: {min_time}")
print(f"latest time: {max_time}")
```

```python
# Proportion of each criterion in DMisinfo Dataset
import pandas as pd

dmisinfo = pd.read_excel("data/manual_annotation.xlsx")
dmisinfo = dmisinfo[dmisinfo["Label"] == 1]

consensus = (dmisinfo["Alignment with Medical Consensus"] < 0).mean()
num_consensus = (dmisinfo["Alignment with Medical Consensus"] < 0).sum()

interventions = (dmisinfo["Information on Interventions"] < 0).mean()
num_interventions = (dmisinfo["Information on Interventions"] < 0).sum()

treatment = (dmisinfo["Evidence-based Treatment"] < 0).mean()
num_treatment = (dmisinfo["Evidence-based Treatment"] < 0).sum()

print(f"Consensus < 0: {num_consensus} {consensus:.1%}")
print(f"Interventions < 0: {num_interventions} {interventions:.1%}")
print(f"Treatment < 0: {num_treatment} {treatment:.1%}")
```

```python
# Number of comments associated with DMisinfo Dataset
import pandas as pd

posts = pd.read_excel("data/manual_annotation.xlsx")
dmisinfo = posts[posts["Label"] == 1]
ndmisinfo = posts[posts["Label"] == 0]

n_dmisinfo = (dmisinfo["Total_comments"]).sum()
n_ndmisinfo = (ndmisinfo["Total_comments"]).sum()

av_comments_misinfo = dmisinfo["Total_comments"].mean()
av_comments_nmisinfo = ndmisinfo["Total_comments"].mean()

av_score_misinfo = dmisinfo["Score"].mean()
av_score_nmisinfo = ndmisinfo["Score"].mean()

print(f"Number of comments associated with DMisinfo:\n{n_dmisinfo}")
print(f"Average number of comments associated with DMisinfo:\n{av_comments_misinfo:.0f}")
print(f"Average number of score associated with DMisinfo:\n{av_score_misinfo:.0f}")
print(f"\nNumber of comments associated with non-DMisinfo:\n{n_ndmisinfo}")
print(f"Average number of comments associated with DMisinfo:\n{av_comments_nmisinfo:.0f}")
print(f"Average number of score associated with DMisinfo:\n{av_score_nmisinfo:.0f}")
```

```python
# Summary of keywords used in data collection process
import pandas as pd
import random

df = pd.read_csv("data/posts_en.csv")

results = []
for keyword, group in df.groupby("Keyword_matched"):
    count = len(group)
    example = random.choice(group["Title"].tolist())
    results.append((keyword, count, example))

print("| Keyword                   | Posts | Example Post Title |")
print("| ------------------------- | ----- | ------------------ |")
for keyword, count, example in sorted(results, key=lambda x: -x[1]):
    print(f"| {keyword:<25} | {count:<5} | {example[:80]}{'...' if len(example) > 80 else ''} |")
```

```python
# Summary for DPMisinfo-Large dataset
import pandas as pd

posts = pd.read_csv("data/labeled_by_roberta_lstm.csv")
dmisinfo = posts[posts["label_pred"] == 1]
nmisinfo = posts[posts["label_pred"] == 0]

dmisinfo_n = len(dmisinfo)
nmisinfo_n = len(nmisinfo)

dmisinfo_c = dmisinfo["Total_comments"].sum()
nmisinfo_c = nmisinfo["Total_comments"].sum()

dmisinfo_s_av = dmisinfo["Score"].mean()
nmisinfo_s_av = nmisinfo["Score"].mean()

dmisinfo_c_av = dmisinfo["Total_comments"].mean()
nmisinfo_c_av = nmisinfo["Total_comments"].mean()

print("Category | Posts | Comments | Av. Score| Av. Comments")
print(f"non-DMisinfo | {nmisinfo_n} | {nmisinfo_c} |{nmisinfo_s_av:.0f} | {nmisinfo_c_av:.0f}")
print(f"DMisinfo | {dmisinfo_n} | {dmisinfo_c} |{dmisinfo_s_av:.0f} | {dmisinfo_c_av:.0f}")
```

```python
# LIWC Summary
import pandas as pd

df = pd.read_excel("reports/liwc_comparison_results.xlsx")

df["non-DMisinfo"] = df["Mean_NonDM"].round(2)
df["DMisinfo"] = df["Mean_DM"].round(2)
df["Diff"] = (df["Diff"] * 100).round(2)
df["d"] = df["Cohen_d"].round(3)

df = df[["Category", "non-DMisinfo", "DMisinfo", "Diff", "stars", "d"]]
print(df.to_markdown(index=False))
```
